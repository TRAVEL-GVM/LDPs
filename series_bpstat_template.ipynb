```python
def transform_data(d):
    new_data = {}
    for key, value in d.items():
        new_data[key] = {}
        for sub_key, url in value.items():
            if "Quarterly" in sub_key:
                periodicity = "Quarterly"
                new_key = sub_key.replace(", Quarterly", "")
            elif "Monthly" in sub_key:
                periodicity = "Monthly"
                new_key = sub_key.replace(", Monthly", "")
            elif "Annual" in sub_key:
                periodicity = "Annual"
                new_key = sub_key.replace(", Annual", "")
            else:
                periodicity = "Unknown"
                new_key = sub_key
            
            new_data[key][new_key] = {
                "periodicity": periodicity,
                "url": url
            }
    
    return new_data

transformed_data = transform_data(indicadores_ecb)
```

```python
def extract_and_combine_data(transformed_data, periodicity, start_date='2020-01'):
    combined_series = pd.DataFrame()

    for category, indicators in transformed_data.items():
        for indicator, details in indicators.items():
            if details['periodicity'] == periodicity:
                url = details['url'].split("datasets/")[1].split('/')[1]
                data = obi.extract_data_from_ecb(url, start_date)
                if not data.empty:
                    data.rename(columns={'OBS_VALUE': indicator}, inplace=True)
                    if combined_series.empty:
                        combined_series = data
                    else:
                        combined_series = pd.merge(combined_series, data, on='TIME_PERIOD', how='outer')

    return combined_series
```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill, Font, Alignment
from openpyxl.utils import get_column_letter

def write_df_to_sheet(sheet, df):
    for c_idx, col_name in enumerate(df.columns, 1):
        sheet.cell(row=1, column=c_idx, value=col_name)
    for r_idx, row in enumerate(df.values, 1):
        for c_idx, value in enumerate(row, 1):
            sheet.cell(row=r_idx + 1, column=c_idx, value=value)

yellow_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")
white_font = Font(color="FFFFFF", bold=True)
green_fill = PatternFill(start_color=obi.nb_default_color1.split('#')[1], end_color=obi.nb_default_color1.split('#')[1], fill_type="solid")
pink_fill = PatternFill(start_color="FFCCE5", end_color="FFCCE5", fill_type="solid")
blue_fill = PatternFill(start_color="311CEE", end_color="311CEE", fill_type="solid")

wb = load_workbook("indicadores_bpstat_com_header.xlsx")

ws1 = wb.active
ws1.title = "BPSTAT_LDPs"

sheet = wb["BPSTAT_LDPs"]

sheet['A3'].fill = blue_fill
sheet['A3'].alignment = Alignment(horizontal="center", vertical="center")
sheet['A3'] = 'Date'
sheet['A3'].font = Font(color="FFFFFF", bold=True) 

for row in sheet.iter_rows(min_row=4, max_row=sheet.max_row, min_col=1, max_col=1):
    for cell in row:
        if isinstance(cell.value, pd.Timestamp):
            cell.value = cell.value.date()
        cell.fill = yellow_fill
        cell.font = Font(color="000000", bold=True) 
        cell.alignment = Alignment(horizontal="center", vertical="center")

for col in range(2, sheet.max_column + 1):
    for row in range(1, 3):
        cell = sheet.cell(row=row, column=col)
        if row == 1:
            cell.fill = pink_fill
            cell.font = Font(color="000000", bold=True)  
        elif row == 2:
            cell.fill = green_fill
            cell.font = Font(color="FFFFFF", bold=True) 
        cell.alignment = Alignment(horizontal="center", vertical="center")

for col in sheet.iter_cols(min_col=1, max_col=sheet.max_column):
    max_length = 0
    for cell in col:
        if cell.value:
            max_length = max(max_length, len(str(cell.value)))
    sheet.column_dimensions[get_column_letter(col[0].column)].width = max_length + 2

ws_quarterly = wb.create_sheet("Quarterly_ECB")
ws_monthly = wb.create_sheet("Monthly_ECB")
ws_annual = wb.create_sheet("Annual_ECB")
ws_readme = wb.create_sheet("README")


write_df_to_sheet(ws_quarterly, quarterly_ecb_data)
write_df_to_sheet(ws_monthly, monthly_ecb_data)
write_df_to_sheet(ws_annual, annual_ecb_data)

for sheet in [ws_quarterly, ws_monthly, ws_annual]:
    for row in sheet.iter_rows(min_row=2, max_row=sheet.max_row, min_col=1, max_col=1):
        for cell in row:
            if isinstance(cell.value, pd.Timestamp):
                cell.value = cell.value.date()
            cell.fill = yellow_fill
            cell.font = Font(color="000000", bold=True)
            cell.alignment = Alignment(horizontal="center", vertical="center")

    for col in range(1, sheet.max_column + 1):
        cell = sheet.cell(row=1, column=col)
        cell.fill = green_fill
        cell.font = Font(color="FFFFFF", bold=True)
        cell.alignment = Alignment(horizontal="center", vertical="center")

    for col in sheet.iter_cols(min_col=1, max_col=sheet.max_column):
        max_length = 0
        for cell in col:
            if cell.value:
                max_length = max(max_length, len(str(cell.value)))
        sheet.column_dimensions[get_column_letter(col[0].column)].width = max_length + 5

    sheet['A1'] = 'Date'
    sheet['A1'].alignment = Alignment(horizontal="center", vertical="center")
    sheet['A1'].font = Font(bold=True, color = "000000")
    sheet['A1'].fill = pink_fill

wb.save("output_final.xlsx")
```
